# 4.collecting more information that will be useful for analysis later on.

**The dataset comprises population and literacy rate data only for the years 2001 and 2011. However, employing the linear interpolation method allows for the derivation of population and literacy rate data for all the remaining years.**

```python
# Finding population for all other years with the help of linear interpolation

# Linear interpolation
start_year = 2001
end_year = 2011

# Calculate the annual change in population
df['Annual Change'] = (df['POPULATION (2011)'] - df['POPULATION (2001)']) / (end_year - start_year)

# Estimate population for intermediate years
intermediate_years = range(2002, 2013)
for year in intermediate_years:
    column_name = f'POPULATION ({year})'
    df[column_name] = (df['POPULATION (2001)'] + df['Annual Change'] * (year - start_year)).round()

# Drop the 'Annual Change' column
df.drop(columns = ['Annual Change'], axis = 1, inplace = True)
```

```python
# Finding literacy rate for all other years with the help of linear interpolation

# Linear interpolation
start_year = 2001
end_year = 2011

# Calculate the annual change in literacy rate
df['Annual Change'] = (df['LITERACY RATE (%) - 2011'] - df['LITERACY RATE (%) - 2001']) / (end_year - start_year)

# Estimate literacy rates for intermediate years
intermediate_years = range(2002, 2013)
for year in intermediate_years:
    column_name = f'LITERACY RATE (%) - {year}'
    df[column_name] = (df['LITERACY RATE (%) - 2001'] + df['Annual Change'] * (year - start_year)).round(2)

# Drop the 'Annual Change' column
df.drop(columns = ['Annual Change'], axis = 1, inplace = True)
```

```python
# Calculating the average population for each state and UT from 2001 to 2012
df['AVERAGE POPULATION (2001-2012)'] = df[['POPULATION (2001)', 'POPULATION (2002)', 'POPULATION (2003)', 'POPULATION (2004)',
                                           'POPULATION (2005)', 'POPULATION (2006)', 'POPULATION (2007)', 'POPULATION (2008)',
                                           'POPULATION (2009)', 'POPULATION (2010)', 'POPULATION (2011)', 'POPULATION (2012)'
                                            ]].mean(axis = 1).round()
```

```python
# Calculating the average literacy rate for each state and UT from 2001 to 2012
df['AVERAGE LITERACY RATE (2001-2012)'] = df[['LITERACY RATE (%) - 2001', 'LITERACY RATE (%) - 2002', 'LITERACY RATE (%) - 2003', 
                                                  'LITERACY RATE (%) - 2004', 'LITERACY RATE (%) - 2005', 'LITERACY RATE (%) - 2006', 
                                                  'LITERACY RATE (%) - 2007', 'LITERACY RATE (%) - 2008', 'LITERACY RATE (%) - 2009', 
                                                  'LITERACY RATE (%) - 2010', 'LITERACY RATE (%) - 2011', 'LITERACY RATE (%) - 2012'
                                                  ]].mean(axis = 1).round(2)
```

```python
# Calculating the population growth rate for each state and UT from 2001 to 2012
df['POPULATION GROWTH RATE (2001-2012)'] = (((df['POPULATION (2012)'] - df['POPULATION (2001)']) / df['POPULATION (2001)']) * 100).round()
```
